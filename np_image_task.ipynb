```python
import cv2
import numpy as np
from PIL import Image

im_array = cv2.imread("ring.png")
display(Image.fromarray(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)))
hsv_image = cv2.cvtColor(im_array, cv2.COLOR_BGR2HSV)

light_red = np.array([0, 50, 50])
dark_red = np.array([10, 255, 255])

mask = cv2.inRange(hsv_image, light_red, dark_red)

im_array[mask>0] = [0,0,0]
display(Image.fromarray(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)))

```
