**Task1**: Create Synthetic dataset of signatures on Pakistani Identity Cards

```python
from faker import Faker
from random import randint 
import pandas as pd
import numpy as np
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display
```

```python
fake_lang = Faker('ar_SA')
```

```python
def synthetic_data(Name):
    img = Image.open('white.jpg')
#     new_img = img.resize((300, 350))
    draw = ImageDraw.Draw(img)
    text = f"{Name}"
    font = ImageFont.truetype("arabicfont1.ttf", 20)
    x = (img.width - 50) // 2
    y = (img.height - 50) // 2
    draw.text((x, y), text, font=font, fill=(0, 0, 0))
    return img

data = []
for _ in range(100):
    Name = fake_lang.name()
    FatherName = fake_lang.name()
    Unique_ID = randint(1, 150)
    Address = fake_lang.address()
    Country = fake_lang.country()
    Signature = synthetic_data(Name)
    Signature.save(f"{Name}.jpg")  # save signature image
    data.append([Name, FatherName, Unique_ID, Address, Country, f"{Name}.jpg"])  # append file path to data

    
    
df = pd.DataFrame(data, columns=['Name', 'FatherName', 'Unique_ID', 'Address','Country', 'Signature'])
df
```

```python
df['Signature'] = df['Signature'].apply(lambda x: Image.open(x))
for index, row in df.iterrows():
    display(row['Signature'])
```
